```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.graphics.tsaplots import plot_pacf,plot_acf
```

ARIMA适用于短期、单变量预测，适用于本问。

```python
endog = [1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
#,
res = [202.73,202.61,207.85,216.26,217.31,232.36,268.13,314.2,386.71,419.5,436.35,462.05,485.97,546.29,
       613.11,619.99,638.44,621.07,634.16,653.12,645.05,644.97,636.58]
alldata = [1997,1998,1999,2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,202.73,202.61,207.85,216.26,217.31,232.36,268.13,314.2,386.71,419.5,436.35,462.05,485.97,546.29,
       613.11,619.99,638.44,621.07,634.16,653.12,645.05,644.97,636.58]
data = np.asarray(alldata).reshape(2,23)
data = data.T
df = pd.DataFrame(data,columns=['time','data']).set_index('time')
plt.plot(df.index,df['data'].values)
plt.show()
```

```python
test = df.copy()
train = df
print(sm.tsa.stattools.adfuller(train['data']))
```

1%、5%、10%不同程度拒绝原假设的统计值和ADF比较，ADF同时小于1%、5%、10%即说明非常好地拒绝该假设，本数据中，adf结果为-3.084， 小于三个level的统计值，说明数据是平稳的.

```python
acorr_ljungbox(train['data'], lags = [6, 12],boxpierce=True)
```

阶延迟下LB和BP统计量的P值都小于显著水平（α = 0.05 \alpha=0.05α=0.05）,所以拒绝序列为纯随机序列的原假设，认为该序列为非白噪声序列。
接下来计算ACF与PACF

```python
acf=plot_acf(train['data'])
plt.title("ACF")
plt.show()
```

```python
pacf=plot_pacf(train['data'],lags=9)
plt.title("PACF")
plt.show()
```

PACF图像有些问题，我们采取AIC和BIC方法来确定阶数（即为参数）
Akaike information criterion（赤池信息量准则）
Bayesian Information Criterion（贝叶斯信息准则）

```python
model=sm.tsa.arima.ARIMA(train,order=(1,1,0))
arima_res = model.fit()
arima_res.summary()
```

```python
# predict = []
# for i in range(1999,2020):
#        predict.append(arima_res.predict(i))
predict = arima_res.predict()
arr = np.asarray(predict)
predict = arr[1:]
plt.plot(test.index,test['data'])
plt.plot(test.index[1:],predict)
plt.show()
print(len(predict))
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
test = test['data'].values
test = test[1:]
print(r2_score(y_true=test,y_pred=predict))
mean_absolute_error(test,predict)
```

残差分析

```python
res=test-predict
residual=list(res)
plt.plot(residual)
print(np.mean(residual))
```

残差正态性检验

```python
import seaborn as sns
from scipy import stats
plt.figure(figsize=(10,5))
ax=plt.subplot(1,2,1)
sns.distplot(residual,fit=stats.norm)
ax=plt.subplot(1,2,2)
res=stats.probplot(residual,plot=plt)
plt.show()
```
