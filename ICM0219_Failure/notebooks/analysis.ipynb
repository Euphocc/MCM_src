```python
import numpy as np
m1 = np.ones((6,6))
m1
```

```python
list = [2,6,8,1/2,1,
          3,4,1/4,1/3,
          1,1/9,1/5,
          1/7,1/6,
          3]
k = 0
for i in range(0,5):
    for j in range(i+1,6):
        m1[i][j] = list[k]
        m1[j][i] = 1/list[k]
        k+=1
m1
```

```python
eigen_value,feature_vectore = np.linalg.eig(m1)
print(eigen_value)
feature_vectore
```

```python
w = feature_vectore[:,0]
w2 = w.real
w2
```

```python
sum =w2.sum()
w2 = w2/sum
w2
```

```python
from sklearn import preprocessing
data = preprocessing.scale(data)
data
```

```python
wi1 = np.matmul(data,w2)
wi1
```
