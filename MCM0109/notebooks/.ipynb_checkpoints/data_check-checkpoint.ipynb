```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../data/txtdata/data.csv', header=None)
# dataarr = data[[6,7,8,15,16,17,24,25,32,33,34,41,43]].values
print(np.any(data.isnull()))
```

```python
# values = data.values
# values = values.astype('float32')
# data = preprocessing.scale(values)
# data = pd.DataFrame(data)
# print(data)
data1 = data.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
res = data1.var()
xdata = range(1,46)
for i in range(len(xdata)):
    plt.bar(xdata[i],res[i])
plt.title("The variance of the feature")
plt.show()
```

```python
data_selected_by_var = [6,7,8,15,16,17,24,25,32,33,34,35,41,42,43,44]
corr = data[data_selected_by_var].corr()
ax=sns.heatmap(corr,vmax=1,vmin=-1)
plt.show()
```

```python
corrarr = corr.values
for i in range(0,16):
    for j in range(0,16):
        if (abs(corrarr[i][j])>=0.78) & (i!=j):
            print(data_selected_by_var[i],data_selected_by_var[j])
```
